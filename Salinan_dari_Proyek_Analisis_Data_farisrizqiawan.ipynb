## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pola penggunaan sepeda di setiap musim?

```python
df_day = pd.DataFrame(day_df)
df_hour = pd.DataFrame(hour_df)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

#Total Pengguna Berdasarkan Day
sns.barplot(ax=axes[0], x="season", y="cnt", data=df_day, palette="coolwarm", ci=None)
axes[0].set_xlabel("Season")
axes[0].set_ylabel("Total Pengguna")
axes[0].set_title("Total Pengguna Berdasarkan Day")
axes[0].set_xticks([0, 1, 2, 3])
axes[0].set_xticklabels(["Spring", "Summer", "Fall", "Winter"])

#Total Pengguna Berdasarkan Hour
sns.barplot(ax=axes[1], x="season", y="cnt", data=df_hour, palette="viridis", ci=None)
axes[1].set_xlabel("Season")
axes[1].set_ylabel("Total Pengguna")
axes[1].set_title("Total Pengguna Berdasarkan Hour")
axes[1].set_xticks([0, 1, 2, 3])
axes[1].set_xticklabels(["Spring", "Summer", "Fall", "Winter"])

plt.tight_layout()
plt.show()
```

### Pertanyaan 2: Bagaimana pola jumlah total peminjaman berdasarkan jumlah pengguna terdaftar?

```python
plt.figure(figsize=(12, 5))
day_grouped = day_df.groupby("registered")["cnt"].sum().reset_index()
hour_grouped = hour_df.groupby("registered")["cnt"].sum().reset_index()

#Total Peminjaman Berdasarkan Day
plt.subplot(1, 2, 1)
sns.barplot(x=day_grouped["registered"], y=day_grouped["cnt"], color="blue")
plt.xlabel("Jumlah Pengguna Terdaftar")
plt.ylabel("Total Peminjaman")
plt.title("Total Peminjaman berdasarkan Registered (Day)")

#Total Peminjaman Berdasarkan Hour
plt.subplot(1, 2, 2)
sns.barplot(x=hour_grouped["registered"], y=hour_grouped["cnt"], color="red")
plt.xlabel("Jumlah Pengguna Terdaftar")
plt.ylabel("Total Peminjaman")
plt.title("Total Peminjaman berdasarkan Registered (Hour)")

plt.tight_layout()
plt.show()
```
